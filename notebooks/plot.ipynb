```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
final_df = pd.read_csv("../data/BTCUSDT_1h_FINAL.csv")
```

```python
def plot_data_split(df, train_limit, val_limit):
    # On s'assure que la colonne date est au bon format
    df['Open_Time'] = pd.to_datetime(df['Open_Time'])
    
    # Création des segments
    train_data = df[df['Open_Time'] < train_limit]
    val_data = df[(df['Open_Time'] >= train_limit) & (df['Open_Time'] < val_limit)]
    test_data = df[df['Open_Time'] >= val_limit]

    # Configuration du style
    plt.figure(figsize=(12, 6))
    plt.style.use('seaborn-v0_8-whitegrid') # Style propre

    # Tracé des lignes
    plt.plot(train_data['Open_Time'], train_data['Close'], color='blue', label='Training', linewidth=1)
    plt.plot(val_data['Open_Time'], val_data['Close'], color='red', label='Validation', linewidth=1)
    plt.plot(test_data['Open_Time'], test_data['Close'], color='green', label='Test', linewidth=1)

    # Personnalisation
    plt.title('BTC/USDT Data Split for DRL Training', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Price (USDT)', fontsize=12)
    plt.legend(loc='upper left', frameon=True)
    
    # Rotation des dates pour plus de lisibilité
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
```

```python
plot_data_split(final_df, '2024-01-01', '2025-01-01')
```
